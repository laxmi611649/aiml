```python
import pandas as pd
import matplotlib.pyplot as plt
iris = pd.read_csv('/content/Iris (1).csv',names=['sepal_length','sepal_width','petal_length','petal_width','class'])
iris.info()
iris.head()
iris.tail()
iris.describe()
import matplotlib.pyplot as plt
plt.figure(figsize=(10,9))
x=iris['petal_length']
y=iris['petal_width']
plt.scatter(x,y)
plt.title("scatter plot petal length vs width")
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.grid(True)
plt.show()
```
